```python
import timeit
import numpy as np
from scipy.linalg import lu, qr, eig, svd
from scipy.linalg.blas import dgemm

# Function to benchmark matrix multiplication
def benchmark_dgemm(size):
    A = np.random.rand(size, size)
    B = np.random.rand(size, size)
    return dgemm(alpha=1.0, a=A, b=B)

# Function to benchmark LU decomposition
def benchmark_lu(size):
    A = np.random.rand(size, size)
    return lu(A)

# Function to benchmark QR decomposition
def benchmark_qr(size):
    A = np.random.rand(size, size)
    return qr(A)

# Function to benchmark Eigenvalue decomposition
def benchmark_eig(size):
    A = np.random.rand(size, size)
    return eig(A)

# Function to benchmark Singular Value Decomposition
def benchmark_svd(size):
    A = np.random.rand(size, size)
    return svd(A)

# Main benchmarking function
def run_benchmarks(size, repeats=3):
    benchmarks = {
        "Matrix Multiplication (dgemm)": "benchmark_dgemm",
        "LU Decomposition (lu)": "benchmark_lu",
        "QR Decomposition (qr)": "benchmark_qr",
        "Eigenvalue Decomposition (eig)": "benchmark_eig",
        "Singular Value Decomposition (svd)": "benchmark_svd"
    }
    
    for name, func in benchmarks.items():
        timer = timeit.Timer(f"{func}({size})", setup=f"from __main__ import {func}")
        time_taken = timer.timeit(number=repeats) / repeats
        print(f"{name}: {time_taken:.6f} seconds")

if __name__ == "__main__":
    matrix_size = 500  # Adjust the matrix size as needed
    run_benchmarks(matrix_size)
```
