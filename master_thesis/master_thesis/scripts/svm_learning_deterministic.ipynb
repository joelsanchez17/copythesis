```python
class Obstacle:
    def __init__(
        self,
        index,
        length,
        height,
        depth,
        center,
        orientation,
        num_points,
        points_radius,
    ):
        self.length = length
        self.index = index
        self.height = height
        self.depth = depth
        self.center = center
        self.orientation = orientation
        self.points_radius = points_radius
        self.num_points = num_points

    def generate_spheres_in_rectangle(
        self,
    ):
        num_points = self.num_points
        length = self.length
        height = self.height
        depth = self.depth
        center = self.center
        orientation = self.orientation
        cx, cy, cz = center
        theta_x, theta_y, theta_z = orientation

        # Generate a grid of points within the 3D rectangle
        x = np.linspace(-length / 2, length / 2, num_points)
        y = np.linspace(-height / 2, height / 2, num_points)
        z = np.linspace(-depth / 2, depth / 2, num_points)
        xv, yv, zv = np.meshgrid(x, y, z, indexing="ij")

        # Flatten the grid and stack the coordinates
        xyz = np.stack((xv.flatten(), yv.flatten(), zv.flatten()), axis=1)

        # Rotate the points based on the orientation angles (assuming radians)
        rotation_matrix_x = np.array(
            [
                [1, 0, 0],
                [0, np.cos(theta_x), -np.sin(theta_x)],
                [0, np.sin(theta_x), np.cos(theta_x)],
            ]
        )

        rotation_matrix_y = np.array(
            [
                [np.cos(theta_y), 0, np.sin(theta_y)],
                [0, 1, 0],
                [-np.sin(theta_y), 0, np.cos(theta_y)],
            ]
        )

        rotation_matrix_z = np.array(
            [
                [np.cos(theta_z), -np.sin(theta_z), 0],
                [np.sin(theta_z), np.cos(theta_z), 0],
                [0, 0, 1],
            ]
        )

        # Combine rotation matrices
        rotation_matrix = rotation_matrix_x @ rotation_matrix_y @ rotation_matrix_z

        # Apply rotation
        rotated_xyz = xyz @ rotation_matrix.T

        # Translate the points to the 3D rectangle's center
        translated_xyz = rotated_xyz + np.array(center)

        return torch.as_tensor(translated_xyz)

    def set_meshcat_objects(
        self,
    ):

        sphere_positions = self.generate_spheres_in_rectangle()

        for i, q_sample in enumerate(sphere_positions):
            meshcat.SetObject(
                f"/obstacles/{self.index}_{i}", Sphere(self.points_radius)
            )
            meshcat.SetProperty(
                f"/obstacles/{self.index}_{i}", "position", q_sample.tolist()
            )


class Environment:
    def __init__(self, robot: GeometricalModel, obstacles):
        self.robot = robot
        self.obstacles = obstacles
        self.obstacle_points = self.calculate_obstacle_points()

    def meshcat_obstacles(
        self,
    ):
        for obstacle in self.obstacles:
            obstacle.set_meshcat_objects()

    def calculate_obstacle_points(self):
        obstacle_positions = []
        obstacle_radii = []
        for obstacle in self.obstacles:
            obstacle_positions.append(obstacle.generate_spheres_in_rectangle())
            obstacle_radii.append(
                torch.tensor(obstacle.points_radius).expand(obstacle.num_points**3, 1)
            )
        return torch.vstack(obstacle_positions), torch.vstack(obstacle_radii)

    def in_collision(self, group, configurations):
        obstacles_point_positions, obstacles_point_radii = (
            self.calculate_obstacle_points()
        )
        # print(obstacles_point_positions)
        return self.robot.is_group_in_collision(
            group, configurations, obstacles_point_positions, obstacles_point_radii
        )


# Example usage

length = 0.2  # Length of the obstacle rectangle
height = 0.1  # Height of the obstacle rectangle
center = (1.5, 0.5)  # Center of the obstacle rectangle

# num_points_robots = 13
meshcat.Delete("/obstacles")

num_points_obstacles = 4
r = 0.04
obstacles = [
    Obstacle(
        0,
        0.5,
        0.1,
        0.1,
        (
            0.0,
            -0.0,
            0.15,
        ),
        (0, 0, 0),
        4,
        r * 2,
    ),
    Obstacle(
        1, 0.3, height, 0.1, (-0.0, 0.5, 0.5), (0, 0, 0), num_points_obstacles, r*2
    ),
    Obstacle(
        2,
        0.3,
        height,
        0.1,
        (-0.25, 0, 0.1),
        (0, np.pi/2, 0),
        num_points_obstacles,
        r * 2,
    ),
]
env = Environment(robots[1].collision_model, obstacles)
env.meshcat_obstacles()
viz_helper.publish_diagram()
collision_model = robots[1].collision_model
# asdfdfs
```

```python
def get_stats(sv_worker):
    collision_model = sv_worker.collision_model
    # obstacle_points, obstacle_radii = env.calculate_obstacle_points()
    # obstacle_points = obstacle_points.to(device = 'cpu',dtype=torch.float32)
    # obstacle_radii = obstacle_radii.to(device = 'cpu',dtype=torch.float32)*0+0.08
    point_cloud = shared_array.copy()
    point_cloud = point_cloud[~np.isnan(point_cloud).any(axis=1)]
    
    point_cloud_size = point_cloud.shape[0]
    obstacle_points = torch.as_tensor(point_cloud).to(device = 'cpu',dtype=torch.float32)
    obstacle_radii = torch.tensor(0.04, dtype=self.dtype, device=self.device).expand(point_cloud_size, 1).to(device = 'cpu',dtype=torch.float32)

    lower_limits = torch.tensor(
        robots[0].plant.GetPositionLowerLimits(), dtype=torch.float32, device='cpu'
    )
    upper_limits = torch.tensor(
        robots[0].plant.GetPositionUpperLimits(), dtype=torch.float32, device='cpu'
    )
    upper_limits[-2:] = 0.001
    lower_limits[-2:] = 0
    uniform_distribution = torch.distributions.uniform.Uniform(lower_limits, upper_limits)

    exploration_samples = uniform_distribution.sample((4000,)).to(
        'cpu'
    ).cpu().to(torch.float64)
    # exploration_samples[:] =  torch.tensor([0.63975457, -0.61873515, -1.08713865, -2.55996813, -0.66606565,
    #     2.12337865, -2.15861275,0.0,0])
    # exploration_samples[:,-2:] = 0
    Y_truth = collision_model.link_collision_modules["all_geometries"].to('cpu') (
        exploration_samples, obstacle_points.cpu().to(torch.float64), obstacle_radii.cpu().to(torch.float64)
    ).reshape(-1, 1)
    Y_pred = torch.zeros_like(Y_truth)
    extra = []
    for group_i, (group_name, group_worker) in enumerate(sv_worker.group_workers.items()):
        # S = group_worker.S_buffer[
        #     : group_worker.num_support_vectors_buffer.item()
        # ].clone().cpu()
        # collision_model.forward_kinematics_groups_torch[group_name] = collision_model.forward_kinematics_groups_torch[group_name].to(device = 'cpu',dtype=torch.float64)
        W = group_worker.W_buffer[: group_worker.num_support_vectors_buffer.item()].clone().to(torch.float64).cpu()
        support_vectors = group_worker.S_buffer[
            : group_worker.num_support_vectors_buffer.item()
        ].clone().cpu()
        # fk_X, K = collision_model.kernel_matrix(group_name, exploration_samples, 'rational_quadratic', alpha, length_scale)
        # group_name,configuration_1,configuration_2, kernel_type, alpha, length_scale
        
        _, _, K = collision_model.partial_forward(
            group_name,
            exploration_samples,
            support_vectors.cpu().to(torch.float64),
            "rational_quadratic",
            alpha_by_group[group_i].cpu().to(torch.float64),
            length_scale_by_group[group_i].cpu().to(torch.float64),
        )
        Y_pred_score = (
            K.cpu().to(torch.float64)
            @ W
        )
        # print("-"*140,Y_pred_score)
        Y_pred |= Y_pred_score > 0
        extra.append((Y_pred_score, group_name))
        # if group_i == 3:
        #     break
    Y_pred_int = Y_pred.int()
    Y_truth_int = Y_truth.int()

    # Calculate True Positives (TP)
    TP = torch.sum((Y_pred_int == 1) & (Y_truth_int == 1))

    # Calculate False Positives (FP)
    FP = torch.sum((Y_pred_int == 1) & (Y_truth_int == 0))

    # Calculate True Negatives (TN)
    TN = torch.sum((Y_pred_int == 0) & (Y_truth_int == 0))

    # Calculate False Negatives (FN)
    FN = torch.sum((Y_pred_int == 0) & (Y_truth_int == 1))

    # Calculate True Negative Rate (TNR) or Specificity
    TNR = TN / (TN + FP).float()

    # Calculate False Negative Rate (FNR)
    FNR = FN / (FN + TP).float()
    # Calculate True Positive Rate (TPR) or Recall
    TPR = TP / (TP + FN)

    # Calculate False Positive Rate (FPR)
    FPR = FP / (FP + TN)

    # Print the results

    return TPR,FPR,TNR,FNR,TP,FP,TN,FN, extra
    # false_positive = exploration_samples[(Y_pred & ~Y_truth).squeeze()]
    # false_negative = exploration_samples[(~Y_pred & Y_truth).squeeze()]
```
